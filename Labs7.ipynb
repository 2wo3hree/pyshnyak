Коэффицент эффективности кода находим по формуле:

$$K_э = \frac{H}{n_{ср}}$$

$$H=-\sum_{i=1}^n p_{i}\cdot \log_2p_i$$

где:
- $H$ - энтропия,
- $p_{i}$ - вероятность $i$-го символа.

$$n_{ср} = \sum_{i=1}^n p_i\cdot n_i$$

где:
- $n_{ср}$ - средняя длина кода,
- $n_i$ - длина кода для $i$-го символа,
- $p_{i}$ - вероятность $i$-го символа.

```python
import math

def efficiency(prop, ln):
  H = 0
  n = 0
  
  for i in prop:
    H -= i * math.log2(i)
    
  for i in range(len(prop)):
    n += prop[i] * ln[i]
    
  return round(H/n, 4)

print(efficiency([0.25, 0.25, 0.125, 0.125, 0.0625, 0.0625, 0.0625, 0.0625], [2,2,3,3,4,4,4,4]))
```

$$D = 1 - \frac{H}{H_{max}} = 0.381$$

где:

*   $H$ - энтропия,
*   $H_{max}$ - максимальная энтропия

```python
def findEntropy(p=[]):
  ans = 0
  for i in p:
    ans -= i*math.log2(i)
    
  return round(ans,4)

p = [0.2, 0.2, 0.2, 0.2]
print(round(1 - (findEntropy(p)/3), 3))
```

1. Поясните назначение и цели эффективного кодирования.
Эффективное кодирование - это процесс перекодировки сообщений с целью более компактной записи, устранения избыточности и уменьшения объема информации на носителе-сигнале. Основная цель эффективного кодирования - обеспечить минимальное число элементов на передачу сообщения источника в среднем. Для этого используются неравномерные коды, в которых более короткие кодовые комбинации соответствуют более вероятным символам сообщения, а более длинные - менее вероятным символам. Основные требования к эффективному коду - однозначность декодирования, минимальное число элементов на передачу сообщения источника в среднем, и отсутствие совпадения начала более длинной комбинации с более короткой.
Эффективное кодирование нашло применение в различных областях, таких как телекоммуникации, компьютерные сети, радиосвязь, криптография и другие. Оно позволяет уменьшить время передачи информации, увеличить ее объем на носителе-сигнале, повысить эффективность системы и обеспечить заданную достоверность при передаче или хранении информации.

Метод Хаффмана обычно более эффективен, чем метод Шеннона-Фано, так как он строит коды символов на основе их частоты появления в сообщении, что позволяет достичь минимальной избыточности. Таким образом, выбор между методами Хаффмана и Шеннона-Фано зависит от конкретной задачи и распределения символов в сообщении.
