Для нахождения пропускной способности канала и скорости передачи информации в канале без шума, мы можем использовать формулы:

1. Пропускная способность канала (C) измеряется в битах в секунду (бит/с) и вычисляется следующим образом:

  $$C = H_{max} / \tau$$

  Где $\tau$ - средняя длительность передачи одного элемента сообщения (0.2 мс), а $H_{max}$ - максимальная энтропия (2 бита)

  $$C = 2 / (0.2 \cdot 10^{-3}) = 10000 \text{ бит/с}$$

2. Скорость передачи информации в канале без шума (V) также измеряется в битах в секунду (бит/с) и зависит от пропускной способности канала и энтропии:

  $$V = - 1/\tau \cdot \sum_{i}p_i\log_2p_i$$
  
  $$V = 5000 \cdot 1.6533 = 8266 \text{ бит/c}$$

Таким образом, пропускная способность канала составляет 10000 бит/с, а скорость передачи информации в канале без шума - 8266 бит/с.

```python
import math

def findEntropy(p=[]):
  ans = 0
  
  for i in p:
    ans -= i * math.log2(i)
    
  return round(ans, 4)

p = [0.34, 0.46, 0.16, 0.04]
print("Энтропия", findEntropy(p))
```

Избыточность источника можно определить сравнивая энтропию источника при статистической независимости символов с энтропией при наличии корреляционных связей между символами.

1. Энтропия при статистической независимости символов $H_0$ определяется по формуле Шеннона:

   $$H_0 = - \sum_ip(xi) * \log_2(p(x_i))$$

   Где $x_i$ - символы алфавита, $p(x_i)$ - вероятности появления символов.

   $$H_0 = - (0.5 * \log_2(0.5) + 0.25 * \log_2(0.25) + 0.125 * \log_2(0.125) + 0.125 * \log_2(0.125))
      \approx - (0.5 * (-1) + 0.25 * (-2) + 0.125 * (-3) + 0.125 * (-3))
      = - (-0.5 - 0.5 - 0.375 - 0.375)
      = 1.75 \text{ бит}$$

2. Теперь рассмотрим энтропию при наличии корреляционных связей между символами. Для этого мы должны использовать условные вероятности.

   $$H_1 = - \sum_i p(x_i) * \sum_j p(x_j|x_i) * \log_2(p(x_j|x_i))$$

   Где $x_i$ и $x_j$ - символы алфавита, $p(x_i)$ - вероятности появления символов, $p(x_j|x_i)$ - условные вероятности появления символа $x_j$ при условии, что предыдущим символом был $x_i$.

   $$H_1 = - (0.5 * (13/16 * \log_2(13/16) + 3/16 * \log_2(3/16)) + 0.25 * (1/8 * \log_2(1/8) + 1/2 * \log_2(1/2) + 3/8 * \log_2(3/8)) + 0.125 * (0 * \log_2(0) + 0 * \log_2(0) + 0 * \log_2(0) + 1 * \log_2(1)) + 0.125 * (1/2 * \log_2(1/2) + 1/4 * \log_2(1/4) + 1/4 * \log_2(1/4)))$$

Вычисляя эти значения, мы получим:

   $$H_1 \approx 1.469\text{ бит}$$

Избыточность источника можно определить как разницу между энтропией при статистической независимости символов и энтропией при зависимости символов:

   Избыточность $= H_0 - H_1 \approx 1.75 - 1.469 = 0.281$ бит

Избыточность источника при статистической независимости символов составляет приближенно 0.281 бит.

1. Найдем энтропию источника сообщений. Энтропия $H$ определяется следующим образом:

   $$H = -\sum_{i=1}^{n} p_i \cdot \log_2(p_i)$$

   где
   - $n$ - количество символов в алфавите (в данном случае 3).
   - $p_i$ - вероятность появления символа $i$ (для $i = 1, 2, 3$).

   В данном случае:
   $$p_1 = 0.2$$
   $$p_2 = 0.7$$
   $$p_3 = 0.1$$

   Подставляем значения и вычисляем:

   $$H = - (0.2 \cdot \log_2(0.2) + 0.7 \cdot \log_2(0.7) + 0.1 \cdot \log_2(0.1)) \approx 1.157 \text{ бит}$$

2. Теперь нам найдем длины кодовых слов для каждого символа алфавита, используя равномерный двоичный код. В данном случае, $n = 3$ (Например, 00, 1, 01), поэтому средняя длина кодового слова будет равна $L = \frac{5}{3} = 1.67$ бит.
3. Пропускная способность канала $C$ вычисляется следующим образом:

   $$C = \text{Частота тактовых импульсов} \cdot \text{Средняя длина кодового слова}$$

   В данном случае, частота тактовых импульсов составляет 500 Гц, а средняя длина кодового слова - 1.67 бита. Подставляем значения:

   $$C = 500 \, \text{Гц} \cdot 1.67 \, \text{бита} = 835 \, \text{бит/сек}$$

4. Скорость передачи информации $V$ равна энтропии источника, деленной на среднюю длину кодового слова:

   $$V = \frac{H}{L}$$
Теперь мы можем вычислить скорость передачи информации:

$$R = \frac{H}{L}$$

$$R = \frac{1.157}{1.67} \approx 0.69$$

Пропускная способность канала $C$ = 1000 бит/сек
Скорость передачи информации $R$ = бит/сек

$$H = -\sum_i^n p(x_i) \cdot \log_2(p(x_i))$$

В данном случае:

$$H = -0,75 \cdot \log_2(0,75) - 0,25 \cdot \log_2(0,25) \approx 0,811 \text{ бит}$$

Избыточность источника:

$$D = 1 - \frac{H}{H_{max}} = 1 - 0.811 = 0.189$$

Теперь определим скорость передачи информации и пропускную способность канала связи.

Скорость передачи информации (с учётом помех):

$$V = \frac{H(X) - H(X|Y)}{\tau}$$

где
- $H(X)$ - энтропия X,
- $H(X|Y)$ - энтропия X при условии помех

Найдем вероятности $x_1$ и $x_2$ с учетом помех:

- $p(x_1|e) = p(x_2)\cdot p(e)$
- $p(x_2|e) = p(x_1)\cdot p(e)$

$$H(X|Y) = \sum_i^n p(x_i|e)\cdot \log_2p(x_i|e) \approx 0.257 \text{ бит}$$

$$V = \frac{H(X) - H(X|Y)}{\tau} = \frac{0.811 - 0.257}{0.1} = 5.54 \text{ бит/с}$$

```python
def findEntropyWithError(p=[], e=1):
  ans = 0
  for i in p:
    ans -= (i*e) * math.log2(i*e)
    
  return round(ans,4)

findEntropyWithError([0.75, 0.25], 0.05)
```

Техническая скорость передачи (или скорость манипуляции) характеризует количество символов, которые могут быть переданы по каналу связи без искажений и задержек. Информационная скорость передачи показывает, сколько полезной информации может быть передано по каналу связи за единицу времени и измеряется в битах в секунду. Для равновероятных сообщений информационная скорость определяется средним количеством информации, которое передается по каналу в единицу времени.
